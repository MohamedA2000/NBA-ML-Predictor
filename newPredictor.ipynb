```python
from sklearn.linear_model import RidgeClassifier
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import TimeSeriesSplit

rr = RidgeClassifier(alpha=1)

split = TimeSeriesSplit (n_splits = 3)
seq = SequentialFeatureSelector (rr, n_features_to_select  =30, direction = "forward", cv = split)
```

```python
removed_columns = ["season", "date", "won", "target", "team", "team_opp"]
```

```python
select_cols = df.columns[~df.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()
df[select_cols] = scale.fit_transform(df[select_cols])
```

```python
df
```

```python
seq.fit(df[select_cols], df ["target"])
```

```python
predictor = list(select_cols[seq.get_support()])
```

```python
predictor
```

```python
def histTest(data, model, predictor, start =2, step =1):
    all_predicts = []
    
    seasons = sorted(data["season"].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i]
        
        trainData = data[data["season"] <season]
        testData =data[data["season"] == season]
        
        model.fit(trainData[predictor], trainData["target"])
        
        pred = model.predict(testData[predictor])
        pred = pd.Series(pred, index = testData.index)
        
        combine = pd.concat([testData["target"], pred], axis =1)
        
        combine.columns = ["actual","predicted"]
        
        all_predicts.append(combine)
    return pd.concat(all_predicts)
```

```python
predictions = histTest(df,rr,predictor)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(predictions["actual"], predictions["predicted"])
```

```python
import numpy as np
```

```python


rolling_avg = df[list(select_cols) + ["won", "team", "season"]]

def average(team):
    numeric_cols = team.select_dtypes(include=np.number).columns
    team[numeric_cols] = team[numeric_cols].rolling(10).mean()
    return team

rolling_avg = rolling_avg.groupby(["team", "season"], group_keys=False).apply(average)
```

```python
rolling_avg
```

```python
rolling_cols = [f"{col}_10" for col in rolling_avg.columns]
rolling_avg.columns = rolling_cols
df = pd.concat([df, rolling_avg], axis=1)
```

```python
df
```

```python
df = df.dropna()
```

```python
df
```

```python
def shift_cols(team, col_name):
    next_col = team[col_name].shift(-1)
    return next_col

def add_cols(df, col_name):
    return df.groupby("team", group_keys=False).apply(lambda x: shift_cols(x,col_name))

df["home_next"] = add_cols(df, "home")
df["team_opp_next"] = add_cols(df, "team_opp")
df["date_next"] = add_cols(df, "date")
```

```python
df
```

```python
df = df.copy()
```

```python
full = df.merge(df[rolling_cols + ["team_opp_next", "date_next", "team"]],
                left_on=["team", "date_next"], 
                right_on=["team_opp_next", "date_next"])
```
